```python
import numpy as np
```

```python
rng = np.random.default_rng(seed=2)
array = rng.integers(low=1, high=100, size=(4, 2))  # 4x2

array
```

```python
np.random.seed(42)
np.random.uniform(0.0, 1.0, size=(2, 3, 1))
```

```python
rng = np.random.default_rng()
array = np.array([1, 2, 3, 4, 5, 6])
rng.shuffle(array)
array
```

```python
fruits = np.array(['apple', 'banana', 'cherry', 'date', 'elderberry'])

fruit = rng.choice(fruits, size=(3, 2))
fruit
```
