```python
import numpy as np
```

```python
# filtering

ages = np.array([
  [21, 17, 19, 20, 16, 18, 35],
  [25, 30, 15, 14, 28, 40, 50]
])

np.shape(ages)
```

```python
underage_mask = ages[ages < 18]
underage_mask
```

```python
adult_mask = ages[(ages >= 18) & (ages < 40)]
adult_mask
```

```python
senior_mask = ages[ages >= 40]
senior_mask
```

```python
adults = np.where((ages >= 18) & (ages < 40), ages, 0)
adults
```
